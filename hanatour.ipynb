```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# 웹 드라이버 경로 설정 (크롬 드라이버 경로를 설정)
driver_path = 'C:/Users/0702s/visualization/chromedriver-win64/chromedriver.exe'

# 옵션 설정 (브라우저를 비활성화된 상태로 실행)
chrome_options = Options()
# chrome_options.add_argument("--headless")  # 브라우저를 보이지 않게 실행
# chrome_options.add_argument("--disable-gpu")

# 웹 드라이버 실행
service = Service(executable_path=driver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# # 1. URL로 이동
url = "https://www.hanatour.com/package/major-products?cntryCd=JP&sort=RPRS_SORT1"
driver.get(url)
```

```python
buttons = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.XPATH, "//a[contains(text(), '판매상품보기')]"))
    )
titles=[]
dates=[]
n=0
while n<5:
    for button in buttons:
        # 버튼 위치까지 화면 스크롤
        driver.execute_script("arguments[0].scrollIntoView(true);", button)
        
        # 버튼 클릭
        driver.execute_script("arguments[0].click();", button)
        print("버튼 클릭됨")
        
        time.sleep(5)  # 상품 정보 로딩 시간 고려
        n=n+1
    print("버튼 다 클릭함")
```

```python
driver.execute_script("window.scrollTo(0,0);")
date_elements=driver.find_elements(By.CSS_SELECTOR, '.calendar ul li a[href="#none"]')
dates=[]
titles=[]
destination_texts=[] #패키지 대분류
periods=[]

#v패키지

destination_elements=driver.find_element(By.CLASS_NAME, 'item_text.eps2')
destination_element=destination_elements.text.strip()

for date_element in date_elements:
    date_text = date_element.find_element(By.XPATH, './span').text.strip()
    if date_text: #날짜 값이 있는지 확인
        driver.execute_script("arguments[0].click();", date_element)

        time.sleep(2)
        print(f"날짜 {date_text} 클릭")

        prod_list_ul = driver.find_element(By.CLASS_NAME, 'type.prod_list_ul')
        
        
        title_elements = prod_list_ul.find_elements(By.CLASS_NAME, 'item_title.eps3')
        period_elements=prod_list_ul.find_elements(By.CSS_SELECTOR, 'p.item_text.air_info span.info')
        
        # 텍스트를 추출
        for title_element, period_element in zip(title_elements, period_elements):
            title_text = title_element.text.strip()
            period_text = period_element.text.strip()
                            
            destination_texts.append(destination_element)
            titles.append(title_text)
            dates.append(date_text)
            periods.append(period_text)
            
        
        print("크롤링 완료")
```

```python

df = pd.DataFrame({'목적지(대분류)': destination_texts,'date': dates, 'title': titles, 'period': periods})
df
```
